I analyzed protein sequences by calculating their lengths, identifying the longest and shortest proteins, and counting amino acid frequencies. This helps understand basic protein structure and composition.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("protein_sequences.csv")
print(df)
```

```python
df["Length"] = df["Sequence"].apply(len)
print(df)
```

```python
print("Longest protein:")
print(df.loc[df["Length"].idxmax()])

print("\nShortest protein:")
print(df.loc[df["Length"].idxmin()])
```

```python
df["Length"].plot(kind='bar')

plt.xlabel("Protein ID")
plt.ylabel("Sequence Length")
plt.title("Protein Sequence Length Distribution")

plt.show()
```

```python
from collections import Counter

all_seq = "".join(df["Sequence"])
freq = Counter(all_seq)

print(freq)
```
